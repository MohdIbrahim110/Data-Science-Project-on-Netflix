# **Data Visualization**

#### **Ploting a bar plot for comparing top 5 countries with most Movies**

```python
movies_by_countries = netflix_data[netflix_data['type'] == 'Movie']
movies_by_countries
```

```python
movies_by_top5_countries = movies_by_countries[['country', 'type']].value_counts().head()
print(type(movies_by_top5_countries))
print(movies_by_top5_countries)
```

```python
movies_by_top5_countries = movies_by_countries[['country', 'type']].value_counts().head().reset_index()
print(type(movies_by_top5_countries))
print(movies_by_top5_countries)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize = (8, 4))
plt.title("Movies by Top 5 Countries")
sns.barplot(movies_by_top5_countries, x = 'country', y = 'count', width = .5)
plt.show()
```

```python
plt.figure(figsize = (8, 4))
plt.title("- TV Shows by Top 5 Countries -")
sns.barplot(tv_shows_by_top5_countries, x = 'country', y = 'count', width = .5)
plt.show()
```
